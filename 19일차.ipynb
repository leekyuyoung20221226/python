# 그룹연산
    1단계 : 분할(split) 데이터를 특정 조건게의해 분할
    2단계 : 적용(apply) 데이터를 집계,변환,필터링하는데 필요한 메소드 적용
    3단계 : 결합(combine) : 2단계의 처리 결과를 하나로 결합

    1단계 : DataFrame객체.groupby(기준이 되는열)

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic = titanic.rename(columns={'sex':'gender'})
```

```python
# titanic[['survived','gender','age','fare','class']]
df = titanic.loc[:,['survived','gender','age','fare','class']]
print('승객수 : ', len(df))
print('-------------')
print(df.head())
grouped = df.groupby(['class'])
print(grouped)
```

```python
df['class'].unique(), df['class'].value_counts()
```

```python
for key,group in grouped:
    print('key : ', key)
    print('group counts : ', len(group))
    print(group.head())
    print('-----------------------------')
```

```python
grouped.mean()
```

```python
# 특정 그룹선택하기
```

```python
grouped.get_group('First')
```

```python
# class, gender 로 분리
```

```python
grouped_two =  df.groupby(['class','gender'])
```

```python
grouped_two
```

```python
# 그룹핑하여 분할한 데이터객체에서 특정 컬럼의집합
grouped.fare.sum()
```

    사용자 정의 함수 적용
    group객체.agg(매핑함수)

```python
# parameter x는 집합을 이루는 Series나 DataFrame 객체
def minMax(x):
    return x.max() - x.min()
```

```python
grouped.agg(minMax)
```

```python
grouped.std()
```

    동시에 여러개의 함수를 매핑
    모든열에 대해서 : group객체.agg([매핑함수, , , ,])
    각각열에 대해서는 :group객체.agg({'열1':매핑함수, '열2':매핑함수 ,,,})

```python
# sum, min, max, mean, minMax -- > 모든 열에 대해서 적용 grouped객체에 적용
```

```python
grouped.agg([sum,min,max,'mean',minMax])
```

```python
# class별로 그룹핑 - 분할:split 한 객체인 grouped
# fare 열에는 min max 적용하고  , age열에는 mean 적용
# agg함수 이용
```

```python
grouped.agg(
    {
        'fare':['min','max'],
        'age' : ['mean']
    }
)
```

    그룹 연산 데이터 변환
    group객체.transform(매핑함수)

```python
# age에 대한 
# z-score :  (각원소 - 평균) / 표준편차
```

```python
# 그룹별 age열의 평균 
age_mean = grouped['age'].mean()
age_mean
```

```python
# 그룹별 age열의 표준편차
age_std = grouped['age'].std()
age_std
```

```python
for key, group in grouped['age']:    
    group_zscore = (group - age_mean.loc[key])/age_std.loc[key]    
    print('origin :', key)
    print(group_zscore.head(2))
    print('---------------------------')
```

```python
def z_score(x):
    return (x - x.mean() ) / x.std()
```

```python
grouped['age'].transform(z_score)
```

    그룹객체를 필터링
    group객체.filter(조건식 함수)

```python
# 선실등급(class)별로 그룹핑된 grouped 객체를 대상으로 데이터 개수가 200개 이상인 그룹을 필터링
```

```python
gruped_filter =  grouped.filter(lambda x : len(x) >= 200)
gruped_filter.head()
```

```python
gruped_filter.describe(include='all')
```

```python
gruped_filter['class'].unique()
```

```python
# 선실등급별 그룹중에서 age 열의 평균이 30보다 작은 그룹
```

```python
age_filter = grouped.filter(lambda x : x.age.mean() < 30)
age_filter.head()
```

    그룹객체에 함수 매핑
    group객체.apply(매핑함수)

```python
# apply는 개별원소에 함수를 일대일로 매핑
```

```python
grouped.apply(lambda x : x.describe())
```

```python
age_apply = grouped.apply(lambda x : x.age.mean() < 30)
```

```python
age_filter.head(2)
```
