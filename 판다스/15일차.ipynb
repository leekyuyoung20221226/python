# 시각화

```python
import matplotlib.pyplot as plt
```

```python
x = range(100)
y = range(100)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x1 = range(100)
y1 = [np.cos(i) for i in x]

x2 = range(100)
y2 = [np.sin(i) for i in x]

plt.plot(x1,y1, label='cos')
plt.plot(x2,y2, label='sin')
plt.legend()
# plt.legend(['cos','sin'])
plt.show()
```

```python
# 축을 설정해서 시각화 - 1
x_list = [x1,x2]
y_list = [y1,y2]
title_list = ['cos','sin']
#화면 분할
fig, ax =  plt.subplots(1,2,figsize=(10,5) )
for idx in range(2):
    ax[idx].plot(x_list[idx], y_list[idx]  )
    ax[idx].set(title = title_list[idx], xlabel = 'x', ylabel = 'y')
plt.show()
```

```python
# 축을 설정해서 시각화 - 2
plt.figure(figsize=(10,5) )
for idx in range(2):
    plt.subplot(1,2,idx+1)
    plt.plot(x_list[idx], y_list[idx]  )
plt.show()    
```

```python
# 2 by 2 화면을 분할
X = np.linspace(-1,1,100)
```

```python
Y = [ 
        [np.sin(X),np.cos(X)],
        [np.tan(X),np.exp(X)] 
]
```

```python
# 축 ax는 행렬
_, ax = plt.subplots(2,2)
for row in range(2):
    for col in range(2):
        ax[row,col].plot(X, Y[row][col])
plt.show()        
        
```

```python
# 축 ax는 행렬
Y = [np.sin(X),np.cos(X),np.tan(X),np.exp(X)]
for idx in range(2*2):
    plt.subplot(2,2, idx+1)
    plt.plot(X, Y[idx])
plt.show()     
```

# 색상

```python
x = range(100)
y = range(100)
plt.plot(x,y,color = 'r')
plt.plot(x,np.array(y)+100, c = '#FFFF00')
plt.show()
```

```python
x = range(100)
y = range(100)
plt.plot(x,y,color = 'r', linestyle = 'dashed')
plt.plot(x,np.array(y)+100, c = '#0FF000', ls ='dotted')
plt.title('two lines')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color = 'r', linestyle = 'dashed')
plt.title('1')

plt.subplot(1,2,2)
plt.plot(x,np.array(y)+100, c = '#0FF000', ls ='dotted')
plt.title('2')
plt.show()
```

```python
x = range(100)
y = range(100)
plt.plot(x,y,color = 'r', linestyle = 'dashed',label = 'line1')
plt.plot(x,np.array(y)+100, c = '#0FF000', ls ='dotted', label = 'line2')
plt.title('two lines')
# plt.legend(loc='best')
plt.legend(loc='upper right')
plt.show()
```

# 막대 그래프 - 범주형 데이터(카테고리)

```python
# 3 * 4
from random import sample
data = sample(range(100), 12)
data = np.array(data).reshape(3,4)
data
```

```python
# x의 시작좌표
x = np.arange(0,8,2)
print(x)
plt.bar(x+0.0, data[0], width = 0.5)
plt.bar(x+0.5, data[1], width = 0.5)
plt.bar(x+1.0, data[2], width = 0.5)
plt.xticks(x+0.5, ('A','B','C','D') )
plt.show()
```

# 누적막대그래프

```python
x = np.arange(data.shape[1])
color_list = ['r','g','b']
data_label = ['A','B','C','D']
```

```python
print(data)
```

```python
for i in range(3):
    print(f"bottom = {np.sum(data[:i], axis=0)}")
    plt.bar(x, data[i], bottom = np.sum(data[:i], axis=0),color = color_list[i], label = data_label[i]  )
plt.legend()
plt.show()
```

# boxplot 4분위수를 시각화 , 데이터의 분포와 밀집 정도를 표현 또는 분산

# 회귀그래프
    머신러닝의 회귀와 괕이 회귀알고리즘을 적용해서 선형(직선)회귀 선을 그래프와 함께 작성

```python
# 새로운 데이터를 로드해서 확인
tips = sns.load_dataset('tips')
# x_ci=95 신뢰구간
sns.regplot(x='total_bill', y='tip',data=tips,x_ci=95)
```

```python
sns.scatterplot(x='total_bill', y='tip',data=tips)
```

```python
# time 별로 분류
sns.scatterplot(x='total_bill', y='tip',data=tips, hue='time')
```

### 비교그래프

```python
sns.countplot(data=tips, x='smoker',  hue = 'day')
plt.show()
```

### 막대그래프
    y값이 연속형 값일 경우 평균을 나타낼때
    데이터의 신뢰 구간을 검은색 막대로  표현

```python
sns.barplot(x='day', y = 'total_bill', data = tips)
plt.show()
```

# 사전정의된 그래프
    여러그래프들을 합쳐서 정보를 추출
    범주형 데이터에 적합

### 바이올린 플롯 : 상자그림과 분포도를 한번에 표시
    x : 범주형 데이터,  y : 연속형 데이터

```python
sns.violinplot(x = 'day',y='total_bill', hue = 'smoker', data=tips)
plt.show()
```

```python
sns.swarmplot(x = 'day',y='total_bill', hue = 'smoker', data=tips)
plt.show()
```

### FaceGrid : 영역을 나눠서 각각 다른 관점에서 범주형 데이터를 시각화
    전체 데이터를 범주형 데이터의 다양한 관점에서 나눠볼 수 있다

```python
g = sns.FacetGrid(data=tips, col = 'time', row = 'day')
g.map(sns.scatterplot,'total_bill', 'tip')
```
