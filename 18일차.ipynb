    데이터 프레임 각 열에 함수 매핑
    DataFrame객체.apply(매핑함수, axis = 0)
    데이터 프레임에 aapply를 적용하면 모든 열을 하나씩 분리해서 매핑함수의 인자로 전달

```python
# 사용자 정의 함수
def missing_value(series):
    return series.isnull()
```

```python
# 데이터프레임의 각열을 인자로 전달하면 Series를반환
result = df.apply(missing_value)
result
```

    
    데이터 프레임객체에 함수 매핑 pip() - 이때 사용하는 함수가 반환하는 리턴값에 따라서 pipe()가 반환하는 리턴타입 결정
    데이터 프레임객체.pip(매핑함수)

```python
# 데이터 프레임을 전달하면 데이터프레임 반환
def missing_value(series):   # true,false
    return series.isnull()
# 데이터 프레임을 전달하면 시리즈를 반환
def missing_count(series):         # 각 열의 NaN갯수
    return missing_value(series).sum()

# 데이터 프레임을 전달하면 값 반환
def total_number_missing(x):       # 총 NaN의 갯수
    return missing_count(x).sum()
```

```python
result_df = df.pipe(missing_value)
result_df.head()
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
df.pipe(total_number_missing)
```
