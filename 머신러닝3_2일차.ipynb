# 모델 훈련

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
```

```python
from random import shuffle
index = list(range(300))
shuffle(index)
fruits = fruits[index]
```

```python
plt.imshow(fruits[0],cmap='gray_r')
plt.axis('off')
plt.show()
```

```python
fruits.shape
```

```python
# 머신러닝 학습을 위해서 2차원으로 변경   dim
fruits_2d = fruits.reshape(300,-1)
```

```python
km = KMeans(n_clusters=3,random_state=0)
km.fit(fruits_2d)
```

```python
km.labels_
```

```python
np.unique(km.labels_, return_counts=True)
```

# 머신러닝이 분류한 첫번째 클러스터(군집)을 출력

```python
predicts = fruits[km.labels_ == 1]
for i in range(len(predicts )):
    plt.subplot(11,11,i+1)
    plt.imshow(predicts[i],cmap='gray_r')
    plt.axis('off')
plt.show()    
```

# 클러스터의 중심값

```python
km.cluster_centers_.shape
```

```python
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(km.cluster_centers_[i].reshape(-1,100), cmap='gray_r')
    plt.axis('off')
    plt.title(i)
plt.show()    
```

```python
# 최적의 k평균은 개수는 엘보우 방식으로 이너셔 값을 찾는다
# 이너셔 inertia - 클러스터에 속한 샘플이 얼마나 가깝게 모여 있는지 나타내는 값
# 이너셔 inertia - 군집이 늘어나면 군집에 속한 셈플이 작아진다. 그래서
# k값을 늘려가변서 이너셔 값이 변화는 구간을 관찰해서 최적의 개수를 찾는다
# 밀집정도를 나타냄 : 이너셔 inertia
km.inertia_
```

```python
inertia = []
for i in range(2,10):
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(fruits_2d)
    inertia.append(km.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(2,10), inertia)
plt.show()
```
