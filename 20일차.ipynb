# 피벗

```python
df = titanic.loc[:,['age','gender','class','fare','survived']]
```

```python
df.head()
```

```python
df1 =pd.pivot_table(
    df          # 피벗할 테이블
    ,index = 'class'
    ,columns= 'gender' 
    ,values = 'age'
    ,aggfunc = 'mean'
)
```

```python
df1
```

```python
df2 =pd.pivot_table(
    df          # 피벗할 테이블
    ,index = 'class'
    ,columns= 'gender' 
    ,values = 'survived'
    ,aggfunc = ['mean','sum']
)
df2
```

```python
# 피벗 
# index class, gender
# column survived
# values age, fare
# aggfunc  mean, max
df3 = pd.pivot_table(
    df
    ,index = ['class','gender']
    ,columns='survived'
    ,values=['age','fare']
    ,aggfunc=['mean','max']
)
df3
```

```python
# gender 인덱스에서 male 값을 갖는 행을 추출
# level 멀티에서 어떤 인덱스 컬럼을 사용할지 설정
# 첫번재 매개변수는 인덱스의 값
df3.xs('female',level='gender')
```

```python
df3.xs(('First','female'))
```

```python
# df2  df에서 survived컬럼을 제외하고 구성
df2 = df.drop(columns=['Survived'])
```

```python
#결측치 age를 채우자..  평균으로 대처
df2.rename(columns={'Sex':'Gender'},inplace=True)
df2.head()
```

```python
# 성별(groupby) 평균 - 분리
gender_grouped = df2.groupby(['Gender'])
```

```python
# 성별 나이 평균 - 적용 - 결합
gender_grouped['Age'].mean()
```

```python
# Pclass(선실등급)별 평균 - 분리
pclass_grouped = df2.groupby(['Pclass'])
```

```python
# 선실등급별 나이 평균 - 적용- 결합
pclass_grouped['Age'].mean()
```

```python
for key, group in pclass_grouped['Age']:
    print('key :', key)
    print('group : ', group.head())
```

```python
fillna_Age = df2['Age'].fillna(
    pclass_grouped['Age'].transform('mean')
)
df3 = pd.DataFrame(df2)
df3['Age'] = fillna_Age
```
