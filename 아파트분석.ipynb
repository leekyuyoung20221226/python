```python
# df['분양가격(㎡)'].astype(int)
df['분양가격(㎡)'] = pd.to_numeric(df['분양가격(㎡)'], errors='coerce')
```

```python
# 데이터 프레임에 평당 가격을 추가
df['평당분양가격'] = df['분양가격(㎡)']*3.3
```

```python
df.head()
```

```python
df['전용면적'] = df['규모구분'].str.replace("전용면적","")
df['전용면적'] = df['전용면적'].str.replace("초과","~")
df['전용면적'] = df['전용면적'].str.replace("이하","")
df['전용면적'] = df['전용면적'].str.replace(" ","")
```

```python
df.head()
```

```python
# 불필요한 컬럼 제거
df = df.drop(columns=['규모구분','분양가격(㎡)'])
```

```python
df.head()
```

```python
# 지역별 평당분양가격
grouped = df.groupby(['지역명'])
grouped['평당분양가격'].mean()
```

```python
# 전용면적 평당분양가격
grouped = df.groupby(['전용면적'])
grouped['평당분양가격'].mean()
```

```python
# 지역명, 전용면적, 평당분양가격의 평균
grouped = df.groupby(['지역명','전용면적'])
grouped['평당분양가격'].mean().unstack().round()
```
