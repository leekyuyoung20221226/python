```python
import gym
env=gym.make('FrozenLake-v1',is_slippery=False)
print(env.observation_space)
print(env.action_space)
n_trial = 20
# 에피소드 수집
env.reset()
episode=[]
for i in range(n_trial):
  action = env.action_space.sample() # 행동(랜덤선택)
  obs,reward,done,info =  env.step(action) # action이 지정한 행동을 수행하고 (새로운 상태,보상,에피소드의 끝, 부가정보) 반환
  episode.append([action,reward,obs])
  # env.render()
  if done:
    break
print(episode)
env.close()
```

```python
# 계산모형
# 마르코프  결정 프로세스(MDP Markov decision process)
# 상태의 종류, 행동의 종류, 보상의 종류를 지정하고, 행동을 취했을때 발생하는 상태변화를 지배하는 규칙
```

```python
# 보상이 주어지는 시점
# 즉시보상 : 다중 손잡이 밴딧
# 지연된 보상 : FrozenLake, 바둑, 장기, 비디오게임 

# 상태집합 : S={s1,s2.... sn}
# 행동집합 : A={a1,a2.... an}
# 보상집합 : R={r1,r2.... rn}

# 밴딧
# 상태집합 : S=0
# 행동집합 : A={손잡이0,,,, 손잡이4}
# 보상집합 : R={1,-1}

# FrozenLake
# 상태집합 : S={0,1,2,3,4,5 .... 15}
# 행동집합 : A={0(left),1,2,3}
# 보상집합 : R={0,1}  목표지점에 도달하면 1이고 나머지는 0

# 상태전이
  # 결정론적 환경(100% 확률로 새로운 상태가 정해지는 환경)
    # 예 :  FrozenLake에서 Right선택하면 100%확률로 오른쪽 이동

  # 스토캐스틱 환경(확률 분포에 따라서 새로운 상태가 다르게 정해지는 환경)
    # env=gym.make('FrozenLake-v1',is_slippery=True)
    # Right를  선택해도 일정한 확률로 오른쪽으로 이동하지 않을수 있음
    # 실제 세계에서 발생하는 바람 또는 얼음위의 미끄러짐을 흉내

```
